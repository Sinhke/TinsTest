```python
def canny_edge(img):
    # Display original image
    cv2.imshow('Original', img)
     
    # Convert to graycsale
    img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
     
    # Canny Edge Detection
    edges = cv2.Canny(image=img_gray, threshold1=1, threshold2=200) # Canny Edge Detection
    # Display Canny Edge Detection Image
    cv2.imshow('Canny Edge Detection', edges)
    cv2.waitKey(0)
```

```python
def blob_detect(img_dir):
    image = cv2.imread(img_dir, cv2.IMREAD_GRAYSCALE)
    
    # Setup SimpleBlobDetector parameters.
    params = cv2.SimpleBlobDetector_Params()
     
    # Change thresholds
    params.minThreshold = 10;
    params.maxThreshold = 200;
     
    # Filter by Area.
    params.filterByArea = True
    params.minArea = 150
     
    # Filter by Circularity
    params.filterByCircularity = True
    params.minCircularity = 0.1
     
    # Filter by Convexity
    params.filterByConvexity = True
    params.minConvexity = 0.87
     
    # Filter by Inertia
    params.filterByInertia = True
    params.minInertiaRatio = 0.01
     
    detector = cv2.SimpleBlobDetector_create(params)
    
    keypoints = detector.detect(image)
    im_with_keypoints = cv2.drawKeypoints(image, keypoints, np.array([]), (0, 0, 255), cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)
    cv2.imshow("BLOBs", im_with_keypoints)
    cv2.waitKey(0)
```

```python
def blob_detect_img(image):
    # Setup SimpleBlobDetector parameters.
    params = cv2.SimpleBlobDetector_Params()
     
    # Change thresholds
    params.minThreshold = 10;
    params.maxThreshold = 200;
     
    # Filter by Area.
    params.filterByArea = True
    params.minArea = 150
     
    # Filter by Circularity
    params.filterByCircularity = True
    params.minCircularity = 0.1
     
    # Filter by Convexity
    params.filterByConvexity = True
    params.minConvexity = 0.87
     
    # Filter by Inertia
    params.filterByInertia = True
    params.minInertiaRatio = 0.01
     
    detector = cv2.SimpleBlobDetector_create(params)
    
    keypoints = detector.detect(image)
    im_with_keypoints = cv2.drawKeypoints(image, keypoints, np.array([]), (0, 0, 255), cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)
    cv2.imshow("BLOBs", im_with_keypoints)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
```

```python
def contour(image):
    # Apply thresholding to isolate bright spots
    _, thresh = cv2.threshold(image, 200, 255, cv2.THRESH_BINARY)
    
    # Find contours of the bright regions
    contours, _ = cv2.findContours(thresh, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    
    # Draw contours on the original image
    for contour in contours:
        if cv2.contourArea(contour) > 50:  # Ignore small spots
            x, y, w, h = cv2.boundingRect(contour)
            cv2.rectangle(image, (x, y), (x + w, y + h), (0, 255, 0), 2)  # Green box

    cv2.imshow("Contour", image)
    cv2.waitKey(0)        
```

```python
def split_img(img_dir):
    image = cv2.imread(img_dir, cv2.IMREAD_GRAYSCALE)

    
    # Get image dimensions
    height, width = image.shape
    
    # Calculate the height of each part
    part_height = height // 3
    
    # Split the image into three parts
    part1 = image[0:part_height, :]
    part2 = image[part_height:2*part_height, :]
    part3 = image[2*part_height:height, :]

    return [part1, part2, part3]
```

```python
def crop_edges(image, crop_x):
    # Get image dimensions
    if len(image.shape) == 3:
        height, width, _ = image.shape
    else:
        height, width = image.shape

    # Ensure X is not larger than half of width or height
    crop_x = min(crop_x, width // 2, height // 2)

    # Crop the image (remove X pixels from each edge)
    cropped_image = image[crop_x:height - crop_x, crop_x:width - crop_x]

    return cropped_image
```

```python
def show_hist(img):
    hist = cv2.calcHist([img], [0], None, [256], [0, 256])
    
    # Plot the histogram
    plt.figure(figsize=(8, 6))
    plt.plot(hist, color='black')
    plt.title("Grayscale Histogram")
    plt.xlabel("Pixel Intensity")
    plt.ylabel("Frequency")
    plt.xlim([0, 256])
    plt.grid()
    plt.show()
```

```python
def non_max_suppression(contours, min_distance=20):
    """ Filters overlapping contours based on centroid distance. """
    filtered_contours = []
    
    for cnt in contours:
        x, y, w, h = cv2.boundingRect(cnt)
        center = (x + w // 2, y + h // 2)
        
        # Check distance from already selected contours
        if all(np.linalg.norm(np.array(center) - np.array(existing_center)) > min_distance for _, existing_center in filtered_contours):
            filtered_contours.append((cnt, center))
    
    return [cnt for cnt, _ in filtered_contours]
```

```python
# Threshold the image to find bright areas
copied_image = adjusted.copy()
_, thresh = cv2.threshold(inverted_image, 150, 255, cv2.THRESH_BINARY)

# Find contours of bright regions
contours, _ = cv2.findContours(thresh, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

# Apply non-max suppression to avoid overlapping
filtered_contours = non_max_suppression(contours, min_distance=30)

# Draw circles around detected bright spots
for cnt in filtered_contours:
    (x, y, w, h) = cv2.boundingRect(cnt)
    center = (x + w // 2, y + h // 2)
    radius = max(w, h) // 2  # Use max dimension for better coverage
    cv2.circle(copied_image, center, radius, (0, 255, 0), 1)  # Green circle

# Create a figure
plt.figure(figsize=(10, 5))

# Show first image
plt.subplot(1, 4, 1)  # (rows, columns, index)
plt.imshow(thresh)
plt.title("Image 1")
plt.axis("off")  # Hide axes

# Show second image
plt.subplot(1, 4, 2)
plt.imshow(inverted_image)
plt.title("Image 2")
plt.axis("off")

# Show first image
plt.subplot(1, 4, 3)  # (rows, columns, index)
plt.imshow(adjusted)
plt.title("Image 1")
plt.axis("off")  # Hide axes

# Show first image
plt.subplot(1, 4, 4)  # (rows, columns, index)
plt.imshow(copied_image)
plt.title("Image 1")
plt.axis("off")  # Hide axes
```

```python
def apply_sharpness(image):
    # Create a sharpening kernel
    sharpen_kernel = np.array([[0, -1, 0],
                               [-1, 5, -1],
                               [0, -1, 0]])
    
    # Apply the kernel using filter2D
    sharpened_image = cv2.filter2D(image, -1, sharpen_kernel)
    return sharpened_image
```

```python
def apply_brightness_contrast(input_img, brightness = 0, contrast = 0):
    
    if brightness != 0:
        if brightness > 0:
            shadow = brightness
            highlight = 255
        else:
            shadow = 0
            highlight = 255 + brightness
        alpha_b = (highlight - shadow)/255
        gamma_b = shadow
        
        buf = cv2.addWeighted(input_img, alpha_b, input_img, 0, gamma_b)
    else:
        buf = input_img.copy()
    
    if contrast != 0:
        f = 131*(contrast + 127)/(127*(131-contrast))
        alpha_c = f
        gamma_c = 127*(1-f)
        
        buf = cv2.addWeighted(buf, alpha_c, buf, 0, gamma_c)

    return buf
```

```python
image_apply = apply_brightness_contrast(image, brightness = 0, contrast = 64)
image_sharpness = apply_sharpness(image)
image_gray = cv2.cvtColor(image_apply, cv2.COLOR_RGB2GRAY)
image_ori_gray = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
image_adjusted = cv2.convertScaleAbs(image, alpha=1.4, beta=30)

# Display results
plt.figure(figsize=(12, 4))

plt.subplot(1, 5, 1)
plt.imshow(image)
plt.title("image")

plt.subplot(1, 5, 2)
plt.imshow(image_apply)
plt.title("apply_brightness_contrast")

plt.subplot(1, 5, 3)
plt.imshow(image_gray)
plt.title("image_gray")

plt.subplot(1, 5, 4)
plt.imshow(image_adjusted)
plt.title("image_adjusted")

plt.subplot(1, 5, 5)
plt.imshow(image_sharpness)
plt.title("image_sharpness")

plt.show()
```
